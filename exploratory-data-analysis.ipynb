```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## **Task 1**

### **i.	Plot a scatter plot for X (random numbers from 0 to 10) and Y (random numbers from 11 to 21). Save the scatter plot**

```python
x = np.arange(0, 10)
y = np.arange(11, 21)
```

```python
print(x)
```

```python
print(y)
```

```python
# plotting scatter plot
plt.scatter(x, y, c='b')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('2D Scatter Plot')
plt.savefig('scatter-plot.png')
```

### **ii.	Create a subplot with different markers and different line colors**

```python
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(2, 2, 2)
plt.plot(x, y, 'g--')

plt.subplot(2, 2, 3)
plt.plot(x, y, 'b*--')

plt.subplot(2, 2, 4)
plt.plot(x, y, 'bo')

plt.show()
```

### **viii.	Plot a scatter plot for Kernel Width vs Length. Write your inference.**

```python
plt.scatter(seeds['Kernel.Length'], seeds['Kernel.Width'], c='b')
plt.xlabel('Kernel Length')
plt.ylabel('Kernel Width')
plt.title('Scatter Plot of Kernel Width vs Kernel Length')
```

Positive correlation

### **ix.	Different type should have different colours in above scatter plot**

```python
sns.scatterplot(x='Kernel.Length', y='Kernel.Width', data=seeds, hue='Type', palette='muted').set(title='Kernel Width vs Kernel Length')
```

Type 3 has less width and length and type 2 has highest. Overlapping can be noted between type 3 and 1 as well as type 1 and 2.

### **x.	Plot a Jointplot to understand relation between Perimeter and Compactness. Write your inference.**

```python
sns.jointplot(x='Perimeter', y='Compactness', data=seeds)
```

Compactness and Perimeter have a positive correlation. Most seeds have perimeter of 13 and 0.88 compactness.

### **xi.	Plot a Box plot to understand correlation between compactness and type.**

```python
seeds['Type'].unique()
```

```python
sns.boxplot(x='Type', y='Compactness', data=seeds).set(title='Boxplot for Compactness vs Type')
```

There are no outliers for type 2 and 3. The compactness in order is type 2, type 1 and lowest being type 3.

### **xii.	Plot a pair plot to understand all characteristics with type being the main parameter. State your inference**

```python
sns.pairplot(seeds, hue='Type', diag_kind='hist', palette='muted')
```

It can be noted that Assymetric coeff. has correlation with other atrributes

### **xiii.	Plot a Violin plot to understand correlation between compactness and type. State your inference**

```python
sns.violinplot(x='Type', y='Compactness', data=seeds, palette='muted').set(title='Violinplot for Compactness vs Type')
```

Type 1 seeds have most compactness around 0.88, min and max being 0.82 and 0.93. For Type 2 the same being 0.88, 0.83, 0.92. Type 3 has the most range of compactness.

### **xiv.	Plot a Kernel Density Estimation plots to understand correlation between compactness and type. State your inference**

```python
sns.FacetGrid(seeds, hue='Type', height=5).map(sns.kdeplot, 'Compactness').add_legend()
```

Density of Type 1 is highest, with Type 2 overlapping it.

### **xv.	Plot a pair plot to understand all characteristics with type being the main parameter. (the main parameter, with KDE instead of histogram in diagonal subplots)**

```python
sns.pairplot(seeds, hue='Type', diag_kind='kde', palette='muted')
```

### **xvi.	An Andrews curve to display separability of data according to Type.**

```python
pd.plotting.andrews_curves(seeds, class_column='Type')
```
